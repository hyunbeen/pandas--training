```python
for i in continent_list:
    df_data = df[df['continent']==i].groupby(["continent","year"])["lifeExp"].mean().to_frame()
    yearlist = [];
    for (continent,year) in df_data.index :
        yearlist.append(year);
    plt.plot(yearlist, df_data.lifeExp, label=i)
    plt.legend(loc='upper left')

plt.title("LifeExp of Year and Continent")
plt.xlabel("year")
plt.ylabel("lifeExp")
plt.show()



```

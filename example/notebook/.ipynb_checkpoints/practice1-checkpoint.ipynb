```python
import pandas as pd
import plotly as plt
df = pd.read_csv('../data\gapminder.tsv', sep = '\t')
```

```python
ax = df_Afghanistan.plot(x='year',y='pop')
```

```python
ax.set_title('Population by Year of Afghanistan')
```

```python
df_Afghanistan = df[df['country']=='Afghanistan']
ax = df_Afghanistan.plot(x='year',y='gdpPercap')
ax.set_title('GdpPerCap by Year of Afghanistan')
```

```python
df_Afghanistan = df[df['country']=='Afghanistan']
ax = df_Afghanistan.plot(x='year',y='lifeExp',kind='bar')
ax.set_title('LifeExp by Year of Afghanistan')
```

```python
df_Afghanistan = df[df['country']=='Afghanistan']
ax = df_Afghanistan.plot.barh(x='year',y='lifeExp')
ax.set_title('LifeExp by Year of Afghanistan')
```

```python
df_Afghanistan = df[df['country']=='Afghanistan']
df_Albania = df[df['country']=='Albania']
df_Austria = df[df['country']=='Austria']
plt.plot(df_Afghanistan["year"], df_Afghanistan["pop"], label='Afghanistan')
plt.plot(df_Albania["year"], df_Albania["pop"], label='Albania')
plt.plot(df_Austria["year"], df_Austria["pop"], label='Austria')
plt.legend(loc='upper left')
plt.title("Population of Year and Continent")
plt.xlabel("year")
plt.ylabel("pop")
plt.show()
```

## 다변량 데이터로 다변량 그래프 그리기 ─ 산점도 그래프(89쪽)

```python
tips = sns.load_dataset('tips')
print(tips.head())
print('-'*60)
```

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
```

```python
#apply() method: 인자로 전달된 함수를 브로드캐스팅 수행 한다.
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head()
#파생변수 : 기존 변수로 새로운 값을 유도한다.
```

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 10,    #size
    c=tips['sex_color'],    #color
    alpha=0.5)              # 투명도

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

## 다양한 종류의 이변량 그래프 그리기(95쪽)

### 1. seaborn 라이브러리로 산점도 그래프 그리기

```python
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

```python
ax = plt.subplots()  #중요
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

```python
jg = sns.jointplot(x='total_bill', y='tip', data=tips) #히스토그램과 산점도를 같이보여줌
jg.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
jg.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") 
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

### 4. 이차원 밀집도 그리기

```python
kde, ax = plt.subplots()  #중요
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'], 
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip') 
```

### 5. 바 그래프 그리기

```python
g =sns.PairGrid(iris, hue="species")
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)

g=g.add_legend()
```

```python
g =sns.PairGrid(iris, hue="species")
g = g.map_diag(plt.hist, histtype="step", linewidth=3)
g = g.map_offdiag(plt.scatter)

g=g.add_legend()
```

```python
g =sns.PairGrid(iris, vars=["sepal_length", "sepal_width"])
g = g.map(plt.scatter)
```

```python
g =sns.PairGrid(iris)
g = g.map_diag(plt.hist, edgecolor="w")
g = g.map_offdiag(plt.scatter, edgecolor="w", s=40)
```

```python
g =sns.PairGrid(
    iris,
    x_vars=["sepal_length", "sepal_width"],
    y_vars=["sepal_length", "sepal_width"]
)
g = g.map(plt.scatter)
```

```python
g =sns.PairGrid(iris)
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot, lw=2)
```

```python
g =sns.PairGrid(
    iris,
    hue="species",
    palette="Set2",
    hue_kws={"marker": ["o", "s", "D"]}
)

g = g.map(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g=g.add_legend()
```

## 그래프에 스타일 적용하기(114쪽)

```python
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```

```python
sns.set_style('whitegrid') 
fig, ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True) 
```

```python
fig = plt.figure() 
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks'] 
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style) 
        
fig.tight_layout() 
```
